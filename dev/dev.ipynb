```python
def getSwapPrice(balance_from,balance_to):
    return (balance_to/balance_from)

def getOutputAmount(amount,balance_from,balance_to):
    return amount * getSwapPrice(balance_from,balance_to)

class dexSimulator:
    def __init__(self):
        self.token1 = 100
        self.token2 = 100
    
    def swapFor2(self,amount):
        output = getOutputAmount(amount,self.token1,self.token2)
        self.token1 += amount
        self.token2 -= output
        return output
    
    def swapFor1(self,amount):
        output = getOutputAmount(amount,self.token2,self.token1)
        self.token2 += amount
        self.token1 -= output
        return output


    def getToken1Price(self):
        return getSwapPrice(self.token1,self.token2)
    
    def getToken2Price(self):
        return getSwapPrice(self.token2,self.token1)
    
    def report(self):
        print("Token 1 Amount: ",self.token1)
        print("Token 2 Amount: ",self.token2)
        print("Token 1 Price: ",self.getToken1Price())
        print("Token 2 Price: ",self.getToken2Price())
        print("Total Balance: ",self.token1 + self.token2)

class Trader:
    def __init__(self):
        self.token1 = 10
        self.token2 = 10

    def getBalance(self):
        print("Token 1 Balance: ",self.token1)
        print("Token 2 Balance: ",self.token2)
        print("Total Balance: ",self.token1 + self.token2)

    def buyToken2(self,dex,amount):
        self.token2 += dex.swapFor2(amount)
        self.token1 -= amount

    def buyToken1(self,dex,amount):
        self.token1 += dex.swapFor1(amount)
        self.token2 -= amount

def reports(dex,trader):
    print("Trader:")
    trader.getBalance()
    print("Dex:")
    dex.report()

```

```python
dex = dexSimulator()
trader = Trader()
trader.buyToken2(dex,10)
reports()
trader.buyToken1(dex,20)
trader.buyToken2(dex,24.44444444446)
```
